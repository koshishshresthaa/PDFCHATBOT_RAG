Handling the documents

```python

from langchain.document_loaders import PyMuPDFLoader
from langchain.embeddings.ollama import OllamaEmbeddings
from langchain.text_splitter import CharacterTextSplitter

class DocumentHandler:
    """
    Handles loading and splitting of documents.
    """
    def __init__(self, file_path):
        """
        Initialize with file paths.
        """
        self.file_path = file_path
        self.documents=[]
        
    def load_documents(self):
        """
        Load documents from file paths.
        """
        loader = PyMuPDFLoader(self.file_path)
        self.documents = loader.load()
        return self.documents


    def split_documents(self, chunk_size=1000, chunk_overlap=200):
        """
        Split documents into smaller chunks.
        """
        splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
        split_docs = splitter.split_documents(self.documents)
        return split_docs
    
    # Function for creating embeddings using FAISS
    def create_embeddings(self,chunks):
        """
        Create vectorstore and return it.
        """
        embedding_model=OllamaEmbeddings(model="llama3")
        # Creating the embeddings using FAISS
        vectorstore = FAISS.from_documents(chunks, embedding_model)
        # returning the vectorstore
        return vectorstore
```

Creating ChatBot Class

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import Ollama

class Chatbot:
    """
    Chatbot that can answer queries from documents and optionally collect user information.
    """
    def __init__(self, ret_chain, enable_user_info=False):
        """
        Initialize with documents retrieval chain and optionally enable user info collection.
        """
#         self.documents = documents
        self.ret_chain=ret_chain
#         self.qa_chain = load_qa_chain(llm=self.llm, chain_type="stuff")
        self.enable_user_info = enable_user_info
        self.user_info = {}

    def handle_query(self, query):
        """
        Handle user query by answering it or collecting user information if requested.
        """
        if self.enable_user_info==True or query.lower()=="call me":
            self._collect_user_info()
            return "Thank you! We will contact you shortly."
        else:
            return self.answer_query(query)

    def answer_query(self, query):
        """
        Answer a query using the documents.
        """
        response= self.ret_chain.invoke({"input": query})
        return response["answer"]

    def _collect_user_info(self):
        """
        Collect user information.
        """
        self.user_info['name'] = input("Please provide your name: ")
        self.user_info['phone'] = input("Please provide your phone number: ")
        self.user_info['email'] = input("Please provide your email: ")
```

Main function to call the chatbot

```python
# !pip install pymupdf
# !pip install faiss-gpu
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain
from langchain.llms import Ollama


from langchain.vectorstores import FAISS

def main():
    #importing llm
    llm= Ollama(model="llama3",temperature=0.1)
    
    
    print("**********Chatbot with PDF Document Support**********")

    # File uploader for PDF documents
    print("\n Upload PDF Documents \n ")
     # Ask user to input PDF path
    pdf_path = input("Enter the path of the PDF document: ")

    if pdf_path:
        # Process uploaded files
        system_prompt = (
            "You are an expert assistant specialized in answering questions based on the content of research papers. "
            "Use the following context extracted from the PDF to provide accurate and comprehensive answers to the user's questions. "
            "If the answer is not present in the context, state that you don't know. "
            "Provide detailed and precise answers to ensure the user receives complete information."
            "\n\n"
            "### PDF Content:\n{context}\n\n"
        )

        prompt = ChatPromptTemplate.from_messages([
            ("system", system_prompt),
            ("human", "{input}"),
        ])

      
        # Load and split documents
        doc_handler = DocumentHandler(pdf_path)
        documents = doc_handler.load_documents()
        split_docs = doc_handler.split_documents()

        # TODO: Create embeddings, vector, and retriever
        #Assume we have the necessary functions for creating embeddings and retriever
        vector = doc_handler.create_embeddings(split_docs)
        retriever = vector.as_retriever()


        document_chain = create_stuff_documents_chain(llm, prompt)
        retrieval_chain = create_retrieval_chain(retriever, document_chain)

        
        while True:
                # Enable user info collection based on user input
                enable_user_info = input("Enable User Info Collection? (yes/no): ").lower() == "yes"

                # Initialize chatbot
                chatbot = Chatbot(retrieval_chain, enable_user_info)
                
                if not enable_user_info:  # If enable_user_info is False
                    query = input("\nAsk a question (or type 'exit' to quit): ")
                    if query.lower() == 'exit':
                        break
                    response = chatbot.handle_query(query)
                    print("\nBot Response:", response)
                else:  # If enable_user_info is True
                    print("\nUser info collection is enabled. No queries will be accepted.")
                    query = None
                    response = chatbot.handle_query(query)
                    print("\nBot Response:", response)
                    
                    # Check if the user wants to exit
                    exit_input = input("Do you want to exit? (yes/no): ").lower()
                    if exit_input == 'yes':
                        break
                        
    # Run the app
if __name__ == "__main__":
    main()
```

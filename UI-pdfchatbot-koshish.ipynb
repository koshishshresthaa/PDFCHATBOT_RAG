Handling the documents

```python
%%writefile documenthandler.py
from langchain.document_loaders import PyMuPDFLoader
from langchain.embeddings.ollama import OllamaEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
class DocumentHandler:
    """
    Handles loading and splitting of documents.
    """
    def __init__(self, file_path):
        """
        Initialize with file paths.
        """
        self.file_path = file_path
        self.documents=[]
        
    def load_documents(self):
        """
        Load documents from file paths.
        """
        loader = PyMuPDFLoader(self.file_path)
        self.documents = loader.load()
        return self.documents


    def split_documents(self, chunk_size=1000, chunk_overlap=200):
        """
        Split documents into smaller chunks.
        """
        splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
        split_docs = splitter.split_documents(self.documents)
        return split_docs
    
    # Function for creating embeddings using FAISS
    def create_embeddings(self,chunks):
        """
        Create vectorstore and return it.
        """
        embedding_model=OllamaEmbeddings(model="llama3")
        # Creating the embeddings using FAISS
        vectorstore = FAISS.from_documents(chunks, embedding_model)
        # returning the vectorstore
        return vectorstore
```

Creating ChatBot Class

```python
%%writefile chatbot.py
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import Ollama
import streamlit as st

class Chatbot:
    """
    Chatbot that can answer queries from documents and optionally collect user information.
    """
    def __init__(self, ret_chain, enable_user_info=False):
        """
        Initialize with documents retrieval chain and optionally enable user info collection.
        """
        self.ret_chain = ret_chain
        self.enable_user_info = enable_user_info
        self.user_info = {}

    def handle_query(self, query):
        """
        Handle user query by answering it or collecting user information if requested.
        """
        if self.enable_user_info or query.lower() == "call me":
            collection_result = self._collect_user_info()

            # If collection is completed, check if all fields are filled
            if collection_result == "All user information collected":
                if all(field for field in self.user_info.values()):
                    return "Thank you! We will contact you shortly."
                else:
                    return "Please fill in all required information."
            # If collection is not completed, return the collection result message
            else:
                return "Collection in progress"
        else:
            return self.answer_query(query)

    def answer_query(self, query):
        """
        Answer a query using the documents.
        """
        response = self.ret_chain.invoke({"input": query})
        return response["answer"]

    def _collect_user_info(self):
        """
        Collect user information.
        """
        if 'name' not in self.user_info:
            self.user_info['name'] = st.text_input("Please provide your name : ", key="name")

        if 'phone' not in self.user_info:
            self.user_info['phone'] = st.text_input("Please provide your phone number: ", key="phone")

        if 'email' not in self.user_info:
            self.user_info['email'] = st.text_input("Please provide your email: ", key="email")

        if st.button("Submit"):
            if all(field for field in self.user_info.values()):
                return "All user information collected"
            else:
                return "User information collection in progress"

```

```python
%%writefile run.py 
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain
from langchain.llms import Ollama
from documenthandler import DocumentHandler
from chatbot import Chatbot
import streamlit as st

# Initialize session state if not already done
if 'step' not in st.session_state:
    st.session_state.step = 0
    st.session_state.enable_user_info = False

def main():
    # Importing llm
    llm = Ollama(model="llama3", temperature=0.1)

    st.title("Chatbot with PDF Document Support")

    # File uploader for PDF documents
    st.sidebar.title("Upload PDF Documents")
    uploaded_file = st.sidebar.file_uploader("Upload a PDF file", type=["pdf"])

    if uploaded_file is not None:
        # Save uploaded file to a path
        pdf_path = uploaded_file.name
        with open(pdf_path, "wb") as f:
            f.write(uploaded_file.getbuffer())

        # Process uploaded files
        system_prompt = (
            "You are an expert assistant specialized in answering questions based on the content of research papers. "
            "Use the following context extracted from the PDF to provide accurate and comprehensive answers to the user's questions. "
            "If the answer is not present in the context, state that you don't know. "
            "Provide detailed and precise answers to ensure the user receives complete information."
            "\n\n"
            "### PDF Content:\n{context}\n\n"
        )

        prompt = ChatPromptTemplate.from_messages([
            ("system", system_prompt),
            ("human", "{input}"),
        ])

        # Load and split documents
        doc_handler = DocumentHandler(pdf_path)
        documents = doc_handler.load_documents()
        split_docs = doc_handler.split_documents()

        # Create embeddings, vector, and retriever
        vector = doc_handler.create_embeddings(split_docs)
        retriever = vector.as_retriever()

        document_chain = create_stuff_documents_chain(llm, prompt)
        retrieval_chain = create_retrieval_chain(retriever, document_chain)

        # Step-by-step UI interaction
        if st.session_state.step == 0:
            enable_user_info_input = st.text_input("Enable User Info Collection? (yes/no): ", key="enable_user_info_input")
            if enable_user_info_input:
                st.session_state.enable_user_info = enable_user_info_input.lower() == "yes"
                st.session_state.step += 1  # Move to the next step

        if st.session_state.step == 1:
            chatbot = Chatbot(retrieval_chain, st.session_state.enable_user_info)
            if st.session_state.enable_user_info:
                st.write("\nUser info collection is enabled. No queries will be accepted.")
                response = chatbot.handle_query(None)
                st.write("\nBot Response:", response)
            else:
                query = st.text_input("Ask a question (or type 'exit' to quit):", key="query_key")
                if query:
                    if query.lower() == 'exit':
                        st.stop()
                    with st.spinner("Generating answer..."):
                        response = chatbot.handle_query(query)
                    st.write("\nBot Response:", response)

        if st.button("Exit"):
            st.stop()

if __name__ == "__main__":
    main()
```
